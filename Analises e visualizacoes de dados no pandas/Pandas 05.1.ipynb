# Resumo e um pouco de visualização de dados com pandas




### Criando nossos dataframes

```python
import pandas as pd

# importar os arquivos
vendas_df = pd.read_csv('Contoso - Vendas - 2017.csv', sep=';', encoding='ISO-8859-1')
produtos_df = pd.read_csv("Contoso - Cadastro Produtos.csv", sep=";", encoding='ISO-8859-1')
clientes_df = pd.read_csv("Contoso - Clientes.csv", sep= ";",  encoding='ISO-8859-1')
lojas_df = pd.read_csv("Contoso - Lojas.csv", sep=";", encoding='ISO-8859-1')


# tratar oagumas colunas 
clientes_df = clientes_df.rename(columns={"ÿID Cliente": "ID Cliente"})
clientes_df = clientes_df.rename(columns={"E-mail": "E-mail Cliente"})
lojas_df = lojas_df.rename(columns={"ÿID Loja": "ID Loja"})
produtos_df = produtos_df.rename(columns={"ÿNome do Produto": "Nome do Produto"})


# pegar somente as colunas que queremos

clientes_df = clientes_df[["ID Cliente", "E-mail Cliente"]]
produtos_df = produtos_df[["ID Produto", "Nome do Produto"]]
lojas_df = lojas_df[["ID Loja", "Nome da Loja"]]


# agora fazemos junção dos DataFrames 

vendas_df = vendas_df.merge(produtos_df, on="ID Produto")
vendas_df = vendas_df.merge(lojas_df, on="ID Loja")
vendas_df = vendas_df.merge(clientes_df, on="ID Cliente")


#exibindo o dataframe final
display(vendas_df)

```

# Qual Cliente que Comprou mais vezes ?
- Usaremos o metudo .value_counts() para contar quantas vezes cada valor do Dataframe aparece.
- Usaremos o metudo .plot() para exebir um gráfico

```python
frequencia_cliente = vendas_df["E-mail Cliente"].value_counts()

display(frequencia_cliente)
# escolher os primeiros 5 clientes
frequencia_cliente[:5].plot(figsize=(17, 5))
```
